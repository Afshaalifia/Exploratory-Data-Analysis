# <b>A. <span style='color:#3A6D8C'>Introduction</span></b>

Exploratory Data Analysis (EDA) adalah proses untuk menganalisis dan memvisualisasikan data secara mendalam sebelum melakukan analisis lebih lanjut atau pemodelan. Tujuannya adalah untuk memahami struktur, pola, dan anomali dalam data. EDA membantu dalam:

* **Mengenali Pola**: Mengidentifikasi hubungan antara variabel dan pola yang mungkin tidak terlihat secara langsung.
* **Menangani Missing Values**: Mengidentifikasi dan menangani data yang hilang atau tidak lengkap.
* **Mendeteksi Outliers**: Mencari nilai ekstrem yang dapat memengaruhi analisis.
* **Visualisasi Data**: Menggunakan grafik dan diagram untuk menyajikan data secara visual, seperti histogram, scatter plot, dan box plot.
* **Statistik Deskriptif**: Menghitung statistik dasar seperti rata-rata, median, dan deviasi standar untuk memberikan gambaran umum tentang distribusi data.

EDA sering kali merupakan langkah pertama dalam proyek analisis data, memberikan wawasan yang penting untuk merumuskan hipotesis dan menentukan langkah selanjutnya.

## <b>D.3. <span style='color:#3A6D8C'>Missing Value</span></b>

Missing value atau nilai yang hilang adalah data yang tidak tersedia dalam dataset.

Penanganan Missing Values:
* `Menghapus`: Menghapus baris atau kolom dengan nilai hilang.
* `Imputasi`: Mengisi nilai hilang dengan nilai lain, seperti rata-rata, median, atau modus.
* `Modeling`: Menggunakan algoritma yang dapat menangani nilai hilang secara langsung.

### <b><span style='color:#55679C'>Quest D.3</span></b>

<i>Dari dataframe `data_sust_melayu`, periksa dan hitung banyaknya missing value pada data, jika ditemukan maka lakukan cara mengatasinya!</i>

```python
# Buat dataframe kosong
data_kosong = pd.DataFrame(index = data_sust_melayu.columns.to_list())

# Lakukan perhitungan
for negara in data_sust_melayu['Country Name'].unique():
    condition = (data_sust_melayu['Country Name'] == negara)
    list_banyak_null = list()
    for col in data_sust_melayu.columns.to_list():
        banyak_nul = data_sust_melayu.loc[condition][col].isna().sum()
        list_banyak_null.append(banyak_nul)
    data_kosong[negara] = list_banyak_null

# Buat highlight merah untuk missing value > 0
styled_df = data_kosong.T.style.map(lambda val : 'background-color: red' if val > 0 else '')

display(styled_df)
```

Ambil contoh misalkan untuk Country Name 'Malaysia' dengan kolom 'Access to electricity (% of population) - EG.ELC.ACCS.ZS'

```python
# Definisikan pengkondisian dan list kolom yang ingin ditampilkan
condition = data_sust_melayu['Country Name'] == 'Malaysia'
list_kolom = ['Country Name', 'Year', 'Access to electricity (% of population) - EG.ELC.ACCS.ZS']

# Tampilkan hasilnya
display(data_sust_melayu.loc[condition, list_kolom])
```

Setelah dilakukan pencarian pada situs

<img src="https://raw.githubusercontent.com/bachtiyarma/Material/refs/heads/main/Image/Materi-Python/Malaysia-Access%20to%20Electricity.png" width="70%">

Terdapat data Access to electricity (% of population) - EG.ELC.ACCS.ZS pada tahun yang nilainya NULL yakni 2000 hingga 2008 dan juga nilai pada tahun 2009 keatas setelah dibandingkan sama, maka lakukan proses imputasi nilai tersebut kedalam dataframe existing dan begitu seterusnya untuk kolom lain

```python
# Definisikan data yang kosong
dict_electricity_malay = {
    2000 : 98.6,
    2001 : 98.6,
    2002 : 98.6,
    2003 : 98.7,
    2004 : 98.7,
    2005 : 98.7,
    2006 : 98.8,
    2007 : 98.9,
    2008 : 99
}

# Definisikan kolom yang akan diimputasi
condition1 = data_sust_melayu['Country Name'] == 'Malaysia'
condition2 = data_sust_melayu['Year'] >= 2000

# Proses imputasi
for tahun in range(2000, 2008 + 1, 1):
    condition2 = data_sust_melayu['Year'] == tahun
    data_sust_melayu.loc[condition1 & condition2, 'Access to electricity (% of population) - EG.ELC.ACCS.ZS'] = dict_electricity_malay.get(tahun)

# Tampilkan hasilnya
list_kolom = ['Country Name', 'Year', 'Access to electricity (% of population) - EG.ELC.ACCS.ZS']

# Tampilkan hasilnya
display(data_sust_melayu.loc[condition, list_kolom])
```

Bagaimana jika data tidak ditemukan? Alih - aling menghapus row data lebih baik data tersebut leave as it is (biarkan null) atau lakukan imputasi dengan berbagai perhitungan seperti imputasi dengan rata-rata, interpolasi atau teknik imputasi lainnya.

## <b>D.4. <span style='color:#3A6D8C'>Outliers</span></b>

Outliers (pencilan) adalah nilai-nilai yang sangat berbeda atau jauh dari nilai-nilai lain dalam suatu dataset. Mereka adalah observasi yang ekstrem atau tidak biasa dibandingkan dengan nilai-nilai lainnya. Outliers bisa menjadi hasil dari kesalahan pengukuran, perbedaan dalam populasi atau sebenarnya, atau mungkin merupakan indikasi dari variasi yang signifikan dalam data.

Outliers dapat diidentifikasi menggunakan metode statistik, seperti menggunakan batas atas dan batas bawah berdasarkan kriteria tertentu, atau dengan memvisualisasikan data menggunakan grafik seperti box plot atau scatter plot. Pendekatan yang umum digunakan adalah dengan menggunakan metode seperti kuartil (interquartile range/IQR) atau tiga kali deviasi standar.
<br><br>

\begin{equation}
\text{Lower Fence} = Q_1 - 1.5 * IQR\\
\text{Upper Fence} = Q_3 + 1.5 * IQR
\end{equation}
<br>

Jika sebuah nilai pada data dibawah nilai Lower Fence disebut *low outlier* sedangkan nilai data diatas Upper Fence disebut *high outlier* <br><br>

### <b><span style='color:#55679C'>Quest D.4</span></b>

<i>Dari dataframe `data_sust_melayu`, periksa outlier pada data, jika ditemukan maka lakukan cara mengatasinya!</i>

```python
# Histogram Sales
import plotly.express as px

number_column = data_sust_melayu.select_dtypes(include = 'number').drop(columns = ['Year'])

col_name = number_column.columns[:5]
color_name = ['#FF8F00', '#AF47D2', '#FFDB00', '#FFAD60', '#7E60BF']

for column, color in zip(col_name, color_name):
    fig = px.box(
        data_sust_melayu,
        x = column,
        color_discrete_sequence  = [color],
    )

    fig.update_traces(
          marker_line_width = 1,
          marker_line_color = 'white'
    )

    fig.update_layout(
        width = 1000,
        height = 600,
        plot_bgcolor = 'rgba(0, 0, 0, 0)',
        title = dict(
            text = f"<b>Box Plot<br><span style='color:{color}; font-size: 15px;'>{column}</b>",
            font = dict(
                size = 28,
                color = '#757882'
            ),
            y = 0.92,
            x = 0.5
        ),
        yaxis = dict(
            title = '',
            showgrid = False,
            showline = False,
            showticklabels = False,
            zeroline = False,
        ),
        margin = dict(
            t = 80,
            b = 10,
            r = 20
        )
    )

    fig.show(renderer = 'colab')
```

Untuk sementara outlier pada data akan dibiarkan saja. Namun kedepan jika akan dilakukan pemodelan maka outlier perlu untuk ditangani. Untuk proses analisis saat ini outlier akan dibiarkan supaya dapat dilihat nilainya pada proses EDA

### <b><span style='color:#55679C'>Quest F.1 - Statistik Deskriptif</span></b>

<i>Hitung statistik deskriptif tiap kolom & berikan interpretasinya!</i>

```python
# Ambil data angka
number_data = data_sust_melayu.select_dtypes(include = 'number').drop(columns = 'Year')

# Hitung statistik deskriptif
statistics = pd.DataFrame({
    'mean' : number_data.mean().round(2),
    'median': number_data.median(),
    'modus': number_data.mode().iloc[0],
    'min': number_data.min(),
    'max': number_data.max(),
    'q1' : number_data.quantile(0.25),
    'q2' : number_data.quantile(0.50),
    'q3' : number_data.quantile(0.75),
    'stdev' : number_data.std().round(2)
})

# Tampilkan hasilnya
display(statistics.T)
```

### <b><span style='color:#55679C'>Quest F.2 - Korelasi</span></b>

<i>Adakah korelasi antar fitur? Buat gambarannya!</i>

```python
# Import library yang dibutuhkan
import plotly.express as px

# Hitung korelasi
korelasi = number_data.corr('pearson')

# Buat Heat Map
fig = px.imshow(
    korelasi,
    text_auto = True,
    color_continuous_scale = 'Blues'
)

# Jangan tampilkan color axes / legend
fig.update_coloraxes(
    showscale = False
)

# Percantik tampilan
fig.update_layout(
    width = 1200,
    height = 1200,
    title = dict(
        text = "<b>Korelasi Antar Variable pada Data</b>",
        font = dict(
            size = 19,
            color = '#0E2954'
        ),
        y = 0.99,
        x = 0.5
    ),
    xaxis = dict(
        tickfont=dict(size=5),
        showticklabels = False
    ),
    yaxis = dict(
        tickfont=dict(size=5),
        showticklabels = False
    ),
    margin = dict(
        t = 10,
        b = 30,
        r = 50,
        l = 50
    )
)

# Tampilkan hasil visualisasi
fig.show(renderer = 'colab')
```
